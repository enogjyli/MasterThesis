```python
def regcoeff(x, y):
    xmean, ymean  = np.mean(x), np.mean(y)
    sxx, sxy = ((x-xmean)**2).sum(), ((x-xmean)*(y-ymean)).sum() 
    b1 = sxy/sxx
    b0 = ymean - b1*xmean
    return b0, b1
```

```python
def forward(bin_len):
    return 2/(bin_len**2+1e-10)

def inverse(gamma_eq):
    return np.sqrt(2/(gamma_eq+1e-10))

def format_func(value, tick_number):
    return f"{value:.2f}"
```
